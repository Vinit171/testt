```python
from librosa import get_duration
from tqdm.notebook import tqdm

clip_durations = {}

for clip in tqdm(clipss):
    duration = get_duration(filename=str(clip))
    #print(duration,"this is duration")

    if duration in clip_durations:
        clip_durations[duration] = clip_durations[duration] + 1
        
    else:
        clip_durations[duration] = 1

clip_durations
```

**3 -  Data Augmentation**

**CLIP SAMPLING FREQUENCY**

```python
from librosa import get_samplerate

clip_sample_rates = {}

for clip in tqdm(clipss):
    sample_rate = get_samplerate(str(clip))

    if sample_rate in clip_sample_rates:
        clip_sample_rates[sample_rate] = clip_sample_rates[sample_rate] + 1
    else:
        clip_sample_rates[sample_rate] = 1
```

```python
clip_sample_rates      # means every clip has 16k frequency
```

```python
table_data = []
table_header = ["#", "Sampling rate", "Count"]

c = 1
for k, v in tqdm(clip_sample_rates.items()):
    table_data.append([c, k, v])
    c += 1
table_data
```
